```python
import numpy as np
pi = np.pi
#from numpy.random import vonmises
from scipy.stats import vonmises
import matplotlib.pyplot as plt
plt.style.use('dark_background')
%matplotlib inline
```

```python
x = np.arange(0, 2*np.pi, 0.01)

ax = plt.subplot(111)

xticks = [0., .5*np.pi, np.pi, 1.5*np.pi, 2*np.pi]
xticks_labels = ["$0$", r"$\frac{\pi}{2}$", r"$\pi$", r"$\frac{3\pi}{2}$", r"$2\pi$"]

ax.set_xticks((xticks), xticks_labels)
ax.set_xticklabels(xticks_labels)

kappas = [0.00001, 1, 5, 20]

for kappa in kappas:
    vm_pdf = vonmises.pdf(x, loc=pi, kappa=kappa)
    ax.plot(x, vm_pdf, label="$\kappa = %.1f$"%kappa)   

    
for spine in plt.gca().spines.values():
    spine.set_visible(False)
    
ax.legend(frameon=False)

plt.savefig('von_mises_demo.png', dpi = 100)
```

```python
x = np.arange(0, 2*np.pi, 0.01)

ax = plt.subplot(111, projection='polar')

xticks = [0., .5*np.pi, np.pi, 1.5*np.pi]
#xticks_labels = ["$0$", r"$\frac{\pi}{2}$", r"$\pi$", r"$\frac{3\pi}{2}$", r"$2\pi$"]

#ax.axis("off")

ax.get_yaxis().set_visible(False)
#ax.get_yaxis().set_visible(False)
ax.get_xaxis()
# for tick in xticks:
#     ax.axvline(tick, color='white', linestyle='dashed')

plt.rc('grid', color='white', linewidth=1, linestyle='dashed')
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)

ax.set_xticks(xticks)
ax.set_xticklabels(xticks_labels)

kappas = [0.00001, 1, 5, 20]

margin = 0 #9.2
border = 0.9 #0.8

for kappa in kappas:
    vm_pdf = vonmises.pdf(x, loc=np.pi/2, kappa=kappa)
    ax.fill_between(x, np.ones(x.shape[0])*(margin+border), vm_pdf+margin+border, alpha=0.8, label='$\kappa=%d$'%kappa)
    #ax.plot(x, vm_pdf+margin+border, alpha=0.8, label='$\kappa=%d$'%kappa)

for spine in plt.gca().spines.values():
    spine.set_visible(False)
    
#ax.legend(frameon=False)

plt.savefig('von_mises_demo_circle.png', dpi = 1000)
```
