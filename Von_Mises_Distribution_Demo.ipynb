```python
import numpy as np
pi = np.pi
#from numpy.random import vonmises
from scipy.stats import vonmises
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0, 2*np.pi, 0.01)

xticks = [0., .5*np.pi, np.pi, 1.5*np.pi, 2*np.pi]
xticks_labels = ["$0$", r"$\frac{\pi}{2}$", r"$\pi$", r"$\frac{3\pi}{2}$", r"$2\pi$"]

plt.xticks((xticks), xticks_labels)

kappas = [0.00001, 1, 5, 20]

for kappa in kappas:
    vm_pdf = vonmises.pdf(x, loc=pi, kappa=kappa)
    plt.plot(x, vm_pdf, label="$\kappa = %.1f$"%kappa)   

    
for spine in plt.gca().spines.values():
    spine.set_visible(False)
    
plt.legend(frameon=False)

plt.savefig('von_mises_demo.png', dpi = 1000)
```
